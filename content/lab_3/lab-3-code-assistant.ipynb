# Introducing Qiskit Code Assistant

As part of IBM's strategy to facilitate the usage of quantum computing, IBM has introduced a generative AI code assistant called **Qiskit Code Assistant**. This is a powerful tool that leverages powerful language models powered by IBM WatsonX, and fine-tuned with extensive Qiskit examples and will help you learn the best ways to write your Qiskit code. As a challenger, we have provided access to you an early release preview version for the Qiskit Code Assistant. This tool is intended to simplify and improve the development process, making quantum computing more accessible and efficient for all users. Qiskit Code Assistant is available via VS Code and Jupyterlab. **For more details, refer to the JupyterLab extension [here](https://pypi.org/project/qiskit-code-assistant-jupyterlab/), the VSCode extension [here](https://marketplace.visualstudio.com/items?itemName=qiskit.qiskit-vscode) and the full paper [here](https://arxiv.org/pdf/2405.19495)**.

<details>
<summary style="font-size: 24px;">Install and run Qiskit Code Assistant on VS Code</summary>

#### Download and Setup Qiskit Code Assistant

- Download [Qiskit Code Assistant](https://marketplace.visualstudio.com/items?itemName=qiskit.qiskit-vscode) from VS Code Extension Marketplace
- Open the [IBM Quantum website](https://quantum.ibm.com/) in your browser and login with your IBM Quantum account. After logging in, a IQP API token is displayed on the upper right side of the webpage.
  


- Press `Cmd/Ctrl+Shift+P` to open the VS Code command palette.
- Type: `Qiskit Code Assistant: Set IBM Quantum API token`.
- Paste your IQP API token in the pop-up dialog and hit `Enter`.



#### Generate code with Qiskit Code Assistant

- The first time you will try to run the Qiskit Code Assistant, a view will show you the disclaimer for the model you're using. Clicking `Accept` will enable the Qiskit Code Assistant to generate the code.
- To generate the code, write out your prompt or function definition with what the function is supposed to do in its docstring and press `Ctrl+.`. This will generate the suggestions in a faded text.



- To accept the suggestion, press `Tab` or `Esc` to reject the suggestion.
- To generate alternate suggestions, press `Ctrl+.` again.
- To cycle forward or backward for the suggested code, press `Alt/Option+]` and `Alt/Option+[`



#### Keyboard chortcuts

The keybindings can be changed in the VSCode `Keyboard Shortcuts` menu for the following commands

| Command | Default shortcut | Description |
|---|---|---|
| `qiskit-code-assistant.accept-suggestion` | `Tab` | Accept the model-generated suggestion |
| `qiskit-code-assistant.handle-get-completion` | `CTRL+.` | Generate suggestions |

</details>

- To cycle forward or backward for the suggested code, press `Alt/Option+]` and `Alt/Option+[`
